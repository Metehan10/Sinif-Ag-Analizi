```python
# Matrisi oluştur
matrix = veri.iloc[:, 3:].values

# Networkx grafiği oluştur
G = nx.DiGraph(matrix)

# Grafik özelliklerini belirle
node_labels = {i: f"{veri.iloc[i, 1]} {veri.iloc[i, 2]}" for i in range(len(veri))}
nx.set_node_attributes(G, node_labels, "Ad")
```

```python
# Grafiği çiz
fig, ax = plt.subplots(figsize=(20,30))
nx.draw_networkx(G, ax=ax, node_shape="o", with_labels=True, labels=node_labels,font_color='black',node_color='blue',edge_color='red')
plt.savefig("grafik.png", format="png")
```

```python
# Düğümler arasındaki yol sayısını hesaplayalım
geodesics = nx.all_pairs_shortest_path_length(G)
# Düğüm kimliklerini indekslere dönüştüren bir sözlük oluşturalım
node_to_index = {node: index for index, node in enumerate(G.nodes)}
# Geodesics Count Matrix'i oluşturalım
num_nodes = len(G.nodes)
matrix = [[0] * num_nodes for _ in range(num_nodes)]

for source, paths in geodesics:
    for target, length in paths.items():
        source_index = node_to_index[source]
        target_index = node_to_index[target]
        matrix[source_index][target_index] = length
        
# Sonucu yazdıralım
for row in matrix:
    print(row)
```

```python
# Otorite değerlerini hesaplamak (örneğin, PageRank kullanarak)
authority_values = nx.pagerank(G)

# Otoriter ağırlığı hesaplamak
authority_weight = {}
for node in G.nodes:
    weight = 0
    for neighbor in G.predecessors(node):
        weight += authority_values[neighbor]
    authority_weight[node] = weight

# Sonucu yazdırmak
for node, weight in authority_weight.items():
    print(f"{node}: {weight}")
```

```python
# Hub değerlerini hesaplamak (HITS kullanarak)
hub_values, _ = nx.hits(G)

# Hub ağırlığını hesaplamak
hub_weight = {}
for node in G.nodes:
    weight = 0
    for neighbor in G.successors(node):
        weight += hub_values[neighbor]
    hub_weight[node] = weight

# Sonucu yazdırmak
for node, weight in hub_weight.items():
    print(f"{node}: {weight}")
```
